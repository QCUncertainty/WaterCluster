```python
import numpy as np
import matplotlib.pyplot as plt

RESULT_FILE = 'trimer_random_100-MBE.out'

N = 3 # no. of monomer in the cluster
MAX_NBODY = 3 # trucation order of the MBE

# find the title line (single string)
tot_e = []
Err = []
with open(RESULT_FILE, 'r') as IN:
    for line in IN:
        line_sp = line.split()
        if len(line_sp) == 1:
            for i in range(2):
                line_read = IN.readline() # get the line of MBE = 2
            for i in range(2, MAX_NBODY):
                line_read_sp = line_read.split()
                tot_e.append(float(line_read_sp[0]))
                line_read = IN.readline()
            line_read_sp = line_read.split()
            tot_e.append(float(line_read_sp[0])) # get the last full body energy
    tot_e_2d = np.reshape(tot_e,(-1, MAX_NBODY - 1))
    for j in range(tot_e_2d.shape[0]):
        for i in range(MAX_NBODY - 2):
            Err.append(float(tot_e_2d[j,i]) - float(tot_e_2d[j,-1]))
    Err = np.transpose(np.reshape(Err,(-1, MAX_NBODY - 2)))

print(Err)
# print(Err[0])

# plot the errors
n_bins = 20
fig, axs = plt.subplots(1, 1, sharey=True, tight_layout=True)
plt.title("Water trimer random-100 B3LYP-D3/6-31G* MBE\n (truncation order = 2) error histogram")
plt.xlabel("Error relative to the full body total energies")
axs.hist(Err[0], bins=n_bins)
```
